```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([[4,2.9,1,0.414],
        [4,4,1,0],
        [1,2.5,-1,0],
        [2.5,1,-1,0.018],
        [4.9,4.5,1,0],
        [1.9,1.9,-1,0],
        [3.5,4,1,0.018],
        [0.5,1.5,-1,0],
        [2,2.1,-1,0.414],
        [4.5,2.5,1,0]])

['r' if itm > 0 else 'b' for itm in data[:,2]]
```

```python
fig,ax = plt.subplots(1,1)
ax.scatter(data[:,0],data[:,1],c=['r' if itm > 0 else 'b' for itm in data[:,2]])
plt.plot()
```

```python
temp = data[data[:,3] > 0,:]
#w = np.sum(data[:,0:1]*data[:,2]*[data[:,3]],axis=0)
w = data[:,0:2].T @ (data[:,2] * data[:,3]).reshape(-1,1)
print(w)
```

```python
alpha_i = data[0,3]
yi = data[0,2]
xi = data[0,0:2]

b = ((alpha_i - alpha_i * yi * (w.T @ xi))/alpha_i * yi)[0]

b =  1/yi - (w.T @ xi)
print(b)
```

```python
# eqn
print(w)
print(b)
```

```python
# take a bunch of x values - calculate y values for these
decision_x_vals = np.linspace(0,5,25)
decision_y_vals = [(-1*w[0]*x + -1*b)/w[1] for x in decision_x_vals]
```

```python
fig,ax = plt.subplots(1,1)
ax.scatter(data[:,0],data[:,1],c=['r' if itm > 0 else 'b' for itm in data[:,2]])
plt.plot(decision_x_vals,decision_y_vals)
plt.plot()
```
